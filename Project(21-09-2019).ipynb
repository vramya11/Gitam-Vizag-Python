# Project on IPL Data set

```python
import pandas as pd
filePath = 'Datafiles/Match.csv'
def readCsvData(filePath):
    return pd.read_csv(filePath)
df = readCsvData(filePath)
print(df)
```

### 1.Display the count of matches based on the year wise

```python
df['Season_Year'].value_counts()
```

### 2.Plot a bar graph  no. of matches won by each team in the year  2011

```python
import matplotlib.pyplot as plt
import seaborn as sns
rows_2011 = df[df.Season_Year == 2011]
sns.countplot(y = rows_2011['match_winner'])
plt.xlabel('winner frequency')
plt.title('win data for the year 2011')
plt.show()
```

### 3.Print  ManOfMatch count of each player in  Hyderabad stadium

```python
Hyderabad = df[df.City_Name == 'Hyderabad']
motm = pd.crosstab(index = Hyderabad['ManOfMach'], columns = 'count',margins = 'false')
print(motm)
```

### 2.Draw Pie plot based on labels as  below Price category and values as count of each category apps  
- labels:

    - Price =0  - Free apps
    - Price <20 - Affordable apps
    - Price<50 - Expensive apps
    - Price >50 - Very expensive
 

```python
lang = ['Free apps','Affordable apps','Expensive apps','Very expensive']
price = [0,20,50,50]
plt.pie(price,labels = lang,shadow=True,startangle=90,autopct='%1.1f%%')
plt.legend(loc='best')
plt.show()
```

```python
movies=['Free apps','Affordable Apps','Expensive apps','Very Expensive']
rating1=df[df.price==0]
rating2=df[(df.price>0) & (df.price<20)]
rating3=df[(df.price>20) & (df.price<50)]
rating4=df[(df.price>50)]
free=rating1.count()[0]
af_ap=rating2.count()[0]
ea= rating3.count()[0]
ve= rating4.count()[0]
li=[free,af_ap,ea,ve]
plt.pie(li,labels=movies,shadow=True,startangle=90)
plt.show()
```
